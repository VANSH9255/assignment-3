```python
def sum_of_even_numbers(numbers):
    """Returns the sum of all even numbers in the list."""
    return sum(num for num in numbers if num % 2 == 0)

# Example usage
if __name__ == "__main__":
    example_list = [1, 2, 3, 4, 5, 6]
    print(f"Sum of even numbers: {sum_of_even_numbers(example_list)}")
```

```python
def reverse_string(s):
    """
    This function takes a string as input and returns its reverse.
    
    Parameters:
    s (str): The string to be reversed.
    
    Returns:
    str: The reversed string.
    """
    return s[::-1]

# Example usage:
original_string = "Hello, world!"
reversed_string = reverse_string(original_string)
print("Original:", original_string)
print("Reversed:", reversed_string)
```

```python
def square_list(numbers):
    """
    This function takes a list of integers and returns a new list containing the squares of each number.
    
    Parameters:
    numbers (list of int): The list of integers to be squared.
    
    Returns:
    list of int: A list containing the squares of the input integers.
    """
    return [x**2 for x in numbers]

# Example usage:
input_list = [1, 2, 3, 4, 5]
squared_list = square_list(input_list)
print("Original list:", input_list)
print("Squared list:", squared_list)
```

```python
def is_prime(number):
    """
    This function checks if a given number is prime.
    
    Parameters:
    number (int): The number to check for primality.
    
    Returns:
    bool: True if the number is prime, False otherwise.
    """
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
```

```python
class FibonacciIterator:
    def __init__(self, num_terms):
        """
        Initialize the Fibonacci iterator with the number of terms.

        Parameters:
        num_terms (int): The number of terms to generate in the Fibonacci sequence.
        """
        self.num_terms = num_terms
        self.current = 0
        self.a, self.b = 0, 1
        self.index = 0

    def __iter__(self):
        """
        Return the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Return the next term in the Fibonacci sequence.
        
        Raises:
        StopIteration: When the number of terms has been generated.
        """
        if self.index >= self.num_terms:
            raise StopIteration
        # Return the current Fibonacci number
        fib_number = self.a
        # Update the Fibonacci numbers
        self.a, self.b = self.b, self.a + self.b
        self.index += 1
        return fib_number

# Example usage:
num_terms = 10
fib_iter = FibonacciIterator(num_terms)

print("Fibonacci sequence:")
for num in fib_iter:
    print(num)
```

```python
def powers_of_2(max_exponent):
    """
    A generator function that yields powers of 2 up to a given exponent.

    Parameters:
    max_exponent (int): The maximum exponent up to which powers of 2 will be generated.

    Yields:
    int: Powers of 2 from 2^0 to 2^max_exponent.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
max_exp = 5
print("Powers of 2 up to 2^{}:".format(max_exp))
for power in powers_of_2(max_exp):
    print(power)
```

```python
def celsius_to_fahrenheit(celsius):
    """
    Convert Celsius temperature to Fahrenheit.
    
    Parameters:
    celsius (float): Temperature in Celsius.
    
    Returns:
    float: Temperature in Fahrenheit.
    """
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map() to convert Celsius to Fahrenheit
fahrenheit_temps = map(celsius_to_fahrenheit, celsius_temps)

# Convert the map object to a list and print the result
fahrenheit_temps_list = list(fahrenheit_temps)
print("Temperatures in Fahrenheit:", fahrenheit_temps_list)
```

```python
def is_not_vowel(char):
    """
    Check if a character is not a vowel.
    
    Parameters:
    char (str): A single character string to check.
    
    Returns:
    bool: True if the character is not a vowel, False otherwise.
    """
    vowels = 'aeiouAEIOU'
    return char not in vowels

def remove_vowels(input_string):
    """
    Remove all vowels from the given string using filter().
    
    Parameters:
    input_string (str): The string from which vowels will be removed.
    
    Returns:
    str: The string with all vowels removed.
    """
    filtered_chars = filter(is_not_vowel, input_string)
    return ''.join(filtered_chars)

# Example usage:
original_string = "Hello, World!"
result_string = remove_vowels(original_string)

print("Original string:", original_string)
print("String without vowels:", result_string)
```

```python
# Sample input: list of sublists
orders = [
    [1, 20.0, 2],  # Order number, price per item, quantity
    [2, 15.0, 5],
    [3, 50.0, 1],
    [4, 80.0, 3],
    [5, 30.0, 4]
]

# Function to calculate adjusted total
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_value = price_per_item * quantity
    if total_value < 100:
        total_value += 10
    return (order_number, total_value)

# Apply the calculation to each order using map and lambda
result = list(map(lambda order: calculate_order(order), orders))

print(result)
```
